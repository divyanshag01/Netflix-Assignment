```python
import pandas as pd
df = pd.read_csv("netflix_customer_churn.csv")
print(df.shape)
print(df.columns)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

numerical_cols = ['age', 'watch_hours', 'monthly_fee', 'churned']
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
categorical_cols = [
    'subscription_type', 'gender', 'region',
    'device', 'payment_method', 'favorite_genre'
]

plt.figure(figsize=(20, 18))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 2, i)
    sns.countplot(data=df, x=col, hue=col, legend=False, order=df[col].value_counts().index, palette='pastel')
    plt.xticks(rotation=45)
    plt.title(f'Count Plot of {col}')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = [
    'age', 'watch_hours', 'last_login_days',
    'monthly_fee', 'number_of_profiles',
    'avg_watch_time_per_day', 'churned'
]

corr = df[num_cols].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

## Insights & Recommendations

1. Customers with lower watch hours tend to have higher churn rates. This suggests that low engagement is a key driver of cancellations.

2. Premium users, despite paying more, show slightly lower churn rates than Basic users. It seems that customers who invest more in the service are more likely to stay.

3. Africa and South America have lower average watch hours compared to other regions. These could be potential focus areas for Netflix to improve engagement.

4. Crypto users have higher churn compared to those using traditional payment methods like Credit or Debit Cards. Maybe Crypto users prefer anonymity or try the service briefly.

5. Tablet users have the highest churn rate among all device types. This might indicate less satisfaction or lower usage on tablets.

6. Action and Comedy genres are the most popular, but Drama has the highest average watch time per day. Netflix could use this to recommend more personalized content.

7. Customers with fewer profiles (1-2) tend to churn more often, which could indicate they're single users not getting as much value.

## Final Thoughts

Focusing on improving user engagement through content recommendations, offering better experience on tablets, and targeting low-engagement regions might help reduce churn.
